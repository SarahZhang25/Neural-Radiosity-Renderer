```python
import mitsuba as mi
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
mi.set_variant('cuda_ad_rgb')  # or 'llvm_ad_rgb' for CPU

def display_mitsuba_scene(fpath):
    scene = mi.load_file(fpath)
    image = mi.render(scene, spp=1024)

    image_array = np.array(image)

    # Convert from linear to sRGB for display
    image_srgb = np.power(np.clip(image_array, 0, 1), 1/2.2)

    # Display the image
    plt.figure(figsize=(10, 10))
    plt.imshow(image_srgb)
    plt.axis('off')
    plt.title('Mitsuba Cube Scene Render')
    plt.tight_layout()
    plt.show()

    # Optionally save the image
    # Image.fromarray((image_srgb * 255).astype(np.uint8)).save('cube_render.png')
```

```python
display_mitsuba_scene('generated_scene.xml')
```

```python
display_mitsuba_scene('scene_one_cube.xml')
# basic test scene with single cube, single light
```
