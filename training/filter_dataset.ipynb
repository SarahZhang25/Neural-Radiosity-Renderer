```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Update this path to your specific renders folder
folder_path = "../data_generation/output_auto/datasets/attempt6_table_chair_540" 
image_paths = glob.glob(os.path.join(folder_path, "*.png"))

corrupted_count = 0
for img_path in image_paths:
    with Image.open(img_path) as img:
        # Convert to RGB to ensure consistent shape
        img_array = np.array(img.convert("RGB"))
        
    # Find perfectly black pixels [0, 0, 0]
    # If compression artifacts exist, you can use a threshold like: np.all(img_array < 5, axis=-1)
    black_pixels = np.all(img_array == 0, axis=-1)
    
    total_pixels = black_pixels.size
    num_black = np.sum(black_pixels)
    ratio = num_black / total_pixels
    
    # if ratio > 0.3:
    #     print(f"Image: {os.path.basename(img_path)} | Black ratio: {ratio:.2%}")
    #     plt.figure(figsize=(4, 4))
    #     # Clip to [0, 1] for HDR display with matplotlib
    #     # plt.imshow(np.clip(img_array, 0, 1))
    #     plt.imshow(img_array)
    #     plt.axis('off')
    #     plt.show()
    # if ratio < 0.3:
        # check if any nans or infs and if so, display
    npz_path = img_path.replace("_render.png", ".npz")
    img_hdr = np.load(npz_path)["hdr_target_image"]
    bad_img = False
    if np.isnan(img_hdr).any():
        # print(f"Image: {os.path.basename(img_path)} has NaNs")
        bad_img = True
    elif np.isinf(img_hdr).any():
        # print(f"Image: {os.path.basename(img_path)} has Infs")
        bad_img = True
    if bad_img:
        # plt.figure(figsize=(4, 4))
        # plt.imshow(img_array)
        # plt.axis('off')
        # plt.show()
        corrupted_count += 1

print(f"Number of corrupted images: {corrupted_count}")
```
